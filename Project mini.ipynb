### Import & Baca CSV

```python
import csv

nilai = [] #angka-angka nilai siswa akan dimasukkan ke []

print("mencoba membaca file...")

try:
    with open("nilai.csv") as file:
        print("File ditemukan!")
        reader = csv.reader(file) #baca isi file baris demi baris
        next(reader) #lewati baris pertama (nama,nilai)

        for row in reader:
            print(row) 
            try:
                nilai.append(int(row[1])) #ambil nilai siswa, ubah ke angka, masukin ke []
            except:
                print("Data tidak valid, dilewati:", row)

except FileNotFoundError:
    print("File tidak ditemukan!")
```

```python
print("Hasil list nilai:", nilai)
```

### Hitung rata-rata, max, min

```python
print("Perhitungan Nilai")

if len(nilai) > 0:
    rata = sum(nilai) / len(nilai) #sum (menjumlahkan), len (banyak siswa), sum/len (rata rata)
    tertinggi = max(nilai) 
    terendah = min(nilai)

    print("Rata-rata :", rata)
    print("Tertinggi :", tertinggi)
    print("Terendah  :", terendah)
else:
    print("Data kosong, tidak bisa menghitung apa-apa.")
```

### Percabangan Sederhana

```python
print("Evaluasi Rata-rata")

if len(nilai) == 0:
    print("Tidak bisa evaluasi karena tidak ada data.")
else:
    if rata >= 80:
        print("Rata-rata bagus!")
    elif rata >= 60:
        print("Rata-rata cukup.")
    else:
        print("Rata-rata rendah.")
```

### Perulangan

```python
print("Daftar Nilai")

for n in nilai:
    print(n)
```

### OOP

```python
class Siswa:
    def __init__(self, nama, nilai):
        self.nama = nama
        try:
            self.nilai = int(nilai)
        except:
            print("Nilai salah, diset 0")
            self.nilai = 0

    def status(self):
        if self.nilai >= 75:
            return "Lulus"
        else:
            return "Tidak lulus"

#contoh
print("Contoh Class Siswa")
contoh = Siswa("Fita", 80)
print(contoh.nama, "-", contoh.nilai, "-", contoh.status())
```

```python
import matplotlib.pyplot as plt

if len(nilai) > 0:
    plt.plot(nilai, color='red', marker='o')
    plt.title("Grafik Nilai Siswa")
    plt.xlabel("Index")
    plt.ylabel("Nilai")
    plt.show()
else:
    print("Grafik tidak bisa dibuat karena tidak ada data.")
```
